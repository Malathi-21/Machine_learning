```python
import pandas as pd
```

```python
df=pd.read_csv('insurance_pre.csv')
df=pd.get_dummies(df,dtype=int,drop_first=True)
df
```

```python
indep=df[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dep=df[['charges']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(indep, dep, test_size=1/3, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.preprocessing import StandardScaler
scy= StandardScaler()      
y_train=scy.fit_transform(y_train.values.reshape(-1,1))
y_test=scy.transform(y_test.values.reshape(-1,1))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
param_grid = {'criterion':['friedman_mse','absolute_error','squared_error'],
              'max_features': [None,'sqrt','log2'],
              'n_estimators':[10,100]} 

grid=GridSearchCV(RandomForestRegressor(),param_grid,refit = True, verbose = 3,n_jobs=-1)
grid.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score
y_pred = grid.predict(x_test)
r2_value = r2_score(y_test, y_pred)
print("The R² score value on the test data is: {:.4f}".format(r2_value))
```

```python
re=grid.cv_results_
print("The R² score value for the best parameters {} is: {:.4f}".format(grid.best_params_, grid.best_score_))

table=pd.DataFrame.from_dict(re)
table
```
