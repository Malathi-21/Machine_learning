```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('insurance_pre.csv')
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
indep=dataset[['age','bmi','children','sex_male','smoker_yes']]
dep=dataset['charges']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(indep,dep,test_size=1/3,random_state=3)
```

MULTILINEARREGRESSION

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
param_grid = {
    'fit_intercept': [True, False],  # Whether to calculate the intercept
    'copy_X': [True, False],         # Whether to copy X before fitting
    'n_jobs': [None, -1],            # Number of jobs to use for computation
    'positive': [True, False] 
}

# Initialize GridSearchCV
grid= GridSearchCV(LinearRegression(),param_grid=param_grid, scoring='neg_mean_squared_error', verbose=3)
   
# fitting the model for grid search 
grid.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score
y_pred = grid.predict(x_test)
r2_value = r2_score(y_test, y_pred)
print("The RÂ² score value on the test data is: {:.4f}".format(r2_value))
```

```python
re=grid.cv_results_

print("The RÂ² score value for the best parameters {} is: {:.4f}".format(grid.best_params_, grid.best_score_))

table=pd.DataFrame.from_dict(re)
table
```

SVR_GRID


```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid = {'kernel':['rbf','sigmoid','linear'],
              'C':[10,100],'gamma':['auto','scale']} 


grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3,n_jobs=-1) 
   
# fitting the model for grid search 
grid.fit(x_train, y_train) #x_trian,y_train
 
```

```python
from sklearn.metrics import r2_score
y_pred = grid.predict(x_test)
r2_value = r2_score(y_test, y_pred)
print("The RÂ² score value on the test data is: {:.4f}".format(r2_value))
```

```python
re=grid.cv_results_

print("The RÂ² score value for the best parameters {} is: {:.4f}".format(grid.best_params_, grid.best_score_))
table=pd.DataFrame.from_dict(re)
table
```

DECISION_TREE_GRID

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
param_grid= {'criterion':['friedman_mse','absolute_error','squared_error','poisson'],
              'max_features': [None,'sqrt','log2'],
              'max_depth': [None, 10, 20, 30],  # Use max_depth instead of max_features
    'min_samples_split': [2, 10, 20]} 
grid = GridSearchCV(DecisionTreeRegressor(), param_grid, refit = True, verbose = 3,n_jobs=-1) 
grid.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score
y_pred = grid.predict(x_test)
r2_value = r2_score(y_test, y_pred)
print("The RÂ² score value on the test data is: {:.4f}".format(r2_value))
```

```python
re=grid.cv_results_
print("The RÂ² score value for the best parameters {} is: {:.4f}".format(grid.best_params_, grid.best_score_))

table=pd.DataFrame.from_dict(re)
table
```

RANDOM_FOREST_GRID

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
param_grid = {'criterion':['friedman_mse','absolute_error','squared_error','poisson'],
              'max_features': [None,'sqrt','log2'],
              'n_estimators':[10,100]} 

grid=GridSearchCV(RandomForestRegressor(),param_grid,refit = True, verbose = 3,n_jobs=-1)
grid.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score
y_pred = grid.predict(x_test)
r2_value = r2_score(y_test, y_pred)
print("The RÂ² score value on the test data is: {:.4f}".format(r2_value))
```

```python
re=grid.cv_results_
print("The RÂ² score value for the best parameters {} is: {:.4f}".format(grid.best_params_, grid.best_score_))

table=pd.DataFrame.from_dict(re)
table
```
